```python
from pandas_datareader.data import Options
from wallstreet import Stock, Call, Put
import pandas as pd
import datetime
```

```python
tickers = pd.read_csv('SP500.csv')
```

```python
now = datetime.datetime.now().strftime('%Y-%m-%d')


#df = pd.DataFrame()
    
def loadPuts(t,expiration):
    
    year=int(expiration.split('-')[2])
    month = int(expiration.split('-')[1])
    day= int(expiration.split('-')[0])

    p = Put(t, y=year, m=month, d=day)

    df = pd.DataFrame.from_dict(p.puts)
    df['ticker'] = t
    df['ticker_price'] = p.underlying.price
    df['date'] = now
    
    return df





#------------------------
i = 0
size = tickers.shape[0]

for t in tickers.ticker[:10]:
    
    i+=1
    
    print('\n\nStarting',t, '--- ', i/size)
    
    try:
        p = Put(t)
    except:
        print('No options for ',t)
    else:
        expirations = p.expirations

        for exp in expirations:
            print('Loading ',t,'for',exp)
            df = pd.concat([df, loadPuts(t, exp)], ignore_index=1)
```

```python
# calculer premium VS strike - prix
# calculer rdt pour une variation de 5%
def rdt(strike,prix,bid,var):
    if (isfloat(strike) & isfloat(bid)):
        return (float(strike) - float(prix)*(1-var))/float(bid)
    
def isfloat(value):
  try:
    float(value)
    return True
  except ValueError:
    return False
```
